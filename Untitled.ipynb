```python
import nltk
import pandas as pd
import numpy as np
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
spam = pd.read_csv("SMSSpamCollection.txt", sep="\t", names=["label", "message"])
```

```python
spam.head()
```

```python
data_set = []
for index,row in spam.iterrows():
    data_set.append((row['message'],row["label"]))
    
```

```python
len(data_set)
```

Preprocessing

```python
stemmer = PorterStemmer()
```

```python
wordnet_Lemma = WordNetLemmatizer()
```

```python
def preprocess(document, stem=True):
    
    #Convert to lower case
    document = document.lower()
    #Tokenize
    words = word_tokenize(document)
    #Remove stop words
    words = [word for word in words if word not in stopwords.words("english")]
    #Stemmer or lemmetise
    if(stem):
        words = [stemmer.stem(word) for word in words]
    else:
        words = [wordnet_Lemma.lemmatize(word, pos='v') for word in words]
        
    document = " ".join(words)

    return document
```

```python
message_set = []
for (message,label) in data_set:
    words_filtered = [word.lower() for word in preprocess(message, stem=False).split() if len(word)>=3]
    message_set.append((words_filtered,label))
```
